```python
import pandas as pd
import string
from collections import Counter

# Load the CSV file into a pandas dataframe (vaccine data)
df = pd.read_csv('/content/drive/My Drive/Misinformation_Project/Bert_classified_impact_score_30K_vaccine_21_22.csv')
```

```python
# Function based on number of followers and friends
# Audience influence score
def AudienceInfluence(n_audience):
    if n_audience > 100000:
        x = 5
    if n_audience > 10000:
        x = 4
    elif n_audience > 1000:
        x = 3
    elif n_audience > 100: 
        x = 2
    elif n_audience > 10: 
        x = 1
    else:
        x = 0
    return x
```

```python
df["Audience Influence"] = (df["Followers"]+df["Friends"]).apply(AudienceInfluence)
df.head()
```

```python
# Save the dataframe to a new CSV file
df.to_csv('/content/drive/My Drive/Misinformation_Project/Audience_Influence_Score_30K_vaccine_21_22.csv', index=False)
```
