# Handling source data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getseries(url,column):
    return pd.read_csv(url).groupby(["Date"])[column].sum()
```

```python
url_interventions="data/classified_impact_score_30K_interventions_21_22.csv"
url_vaccine="data/classified_impact_score_30K_vaccine_21_22.csv"
column1="Impact Value"
column2="Overall Impact"
intervention1 = getseries(url_interventions,column1)
intervention2 = getseries(url_interventions,column2)
vaccine1 = getseries(url_vaccine,column1)
vaccine2 = getseries(url_vaccine,column2)
```

```python
plt.plot(intervention1)
```

```python
plt.plot(intervention2)
```

```python
plt.plot(vaccine1)
```

```python
plt.plot(vaccine2)
```
